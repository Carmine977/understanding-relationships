## Factors influencing extramarital relationships

```python
import statsmodels.api as sm
affairs_df = sm.datasets.fair.load_pandas().data
affairs_df.head()
```

```python
affairs_df['affair_binary'] = affairs_df['affairs']>0
```

```python
# Correlation matrix to highlight variables linearly correlated  
affairs_df.corr()
```

```python
import seaborn as sns
sns.heatmap(affairs_df.corr())
```

We can use a decision tree to identify variables that are correlated to each other (not only linearly)

```python
affairs_X = affairs_df.drop(['affairs', 'affair_binary'], axis=1)
affairs_y = affairs_df['affair_binary']
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
```

```python
model = DecisionTreeClassifier()
scores = cross_val_score(model, affairs_X, affairs_y, cv=10)
```

```python
# average acuracy
print('average: {}'.format(scores.mean()))
# standard deviation
print('stdev: {}'.format(scores.std()))
```

stdev is low. This confirms that the model behaves similarly for each fold.  
Since model is reliable we can train it on the entire dataset.

```python
# Fit model
import pandas as pd
model.fit(affairs_X, affairs_y)
pd.DataFrame({'feature': affairs_X.columns, 'importance': model.feature_importances_}).sort_values('importance').tail(3)
```

Since *occupation_husb* is nominal, the above result doesn't make too much sense.  
Let's convert the variables about occupation to dummy variables.

```python
# manage wife occupation
occupation_dummies = pd.get_dummies(affairs_df['occupation'], prefix='occ_').iloc[:, 1:]
affair_df = pd.concat([affairs_df, occupation_dummies], axis=1)
affair_df.head()
```

```python
# manage husband occupation
occupation_dummies = pd.get_dummies(affairs_df['occupation_husb'], prefix='occ_husb_').iloc[:, 1:]
affairs_df = pd.concat([affairs_df, occupation_dummies], axis=1)
affairs_df.head()
```

```python
affairs_df.shape
```

```python
affairs_X = affairs_df.drop(['affairs', 'affair_binary', 'occupation', 'occupation_husb'], axis=1)
affairs_y = affairs_df['affair_binary']
```

```python
model = DecisionTreeClassifier()
scores = cross_val_score(model, affairs_X, affairs_y, cv=10)
# average acuracy
print('average: {}'.format(scores.mean()))
# standard deviation
print('stdev: {}'.format(scores.std()))
```

```python
# Fit model
import pandas as pd
model.fit(affairs_X, affairs_y)
pd.DataFrame({'feature': affairs_X.columns, 'importance': model.feature_importances_}).sort_values('importance').tail(10)
```
